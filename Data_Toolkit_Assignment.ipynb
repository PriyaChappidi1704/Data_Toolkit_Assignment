# **Data Toolkit Assignment**

1. What is NumPy, and why is it widely used in Python?
  - NumPy is a Python library used for numerical and scientific computing.
It provides efficient multi-dimensional arrays and functions for fast mathematical operations.
2. How does broadcasting work in NumPy?
  - Broadcasting allows NumPy to perform operations on arrays of different shapes.It automatically stretches smaller arrays to match larger ones during arithmetic operations.
3. What is a Pandas DataFrame?
  - A Pandas DataFrame is a 2D labeled data structure with rows and columns.It is similar to a table in SQL or Excel and allows easy data manipulation and analysis.
4.  Explain the use of the groupby() method in Pandas.
  - The groupby() method groups data based on column values. It's used for performing operations like sum, mean, or count on grouped data.
5. Why is Seaborn preferred for statistical visualizations?
  - Seaborn offers built-in themes and functions for complex statistical plots.It integrates well with Pandas and simplifies the creation of attractive graphs.
6. What are the differences between NumPy arrays and Python lists?
  - NumPy arrays are more memory-efficient and faster for numerical operations.Unlike lists, they support vectorized operations and multi-dimensional data.
7. What is a heatmap, and when should it be used?
  - A heatmap displays data in a matrix form using color gradients. It's used to visualize correlation, patterns, or intensity in data.
8. What does the term “vectorized operation” mean in NumPy?
  - Vectorized operations apply functions directly on arrays without loops. They are faster and more concise than traditional Python loops.
9. How does Matplotlib differ from Plotly?
  - Matplotlib is static and great for basic plotting, while Plotly supports interactive charts. Plotly is better for dashboards and web-based visualizations.
10. What is the significance of hierarchical indexing in Pandas?
  - Hierarchical indexing allows multiple levels of indexing in rows or columns. It enables complex data selection and reshaping operations.
11. What is the role of Seaborn's pairplot() function?
  - pairplot() creates scatter plots between all numerical feature combinations. It helps visualize relationships and distributions in a dataset quickly.
12. What is the purpose of the describe() function in Pandas?
  - The describe() function gives summary statistics for numerical columns. It includes count, mean, std deviation, min, max, and percentiles.
13. Why is handling missing data important in Pandas?
  - Missing data can lead to incorrect analysis and results.
Pandas provides tools to detect, remove, or fill these gaps efficiently.
14. What are the benefits of using Plotly for data visualization?
  - Plotly provides interactive, zoomable, and customizable charts.
It is ideal for presentations and dashboards with dynamic data.
15. How does NumPy handle multidimensional arrays?
  - NumPy supports n-dimensional arrays using the ndarray object.
It offers functions for reshaping, indexing, and broadcasting on them.
16. What is the role of Bokeh in data visualization?
  - Bokeh is a Python library for creating interactive web-based visualizations.
It allows real-time updates and complex layouts for big data.
17. Explain the difference between apply() and map() in Pandas.
  - map() works only on Series and is used for element-wise transformations.
apply() works on both Series and DataFrames, applying functions to rows or columns.
18. What are some advanced features of NumPy?
  - NumPy supports broadcasting, masked arrays, linear algebra, and FFT.
It also allows memory-mapped files and C-level performance for operations.
19. How does Pandas simplify time series analysis?
  - Pandas supports time indexing, resampling, and date-range generation.
It makes time-based data analysis and visualization much easier.
20. What is the role of a pivot table in Pandas?
  - A pivot table reshapes and summarizes data using aggregation.
It's useful for analyzing relationships between variables.
21. Why is NumPy's array slicing faster than Python's list slicing?
  - NumPy arrays are stored in contiguous memory blocks.
This allows faster access and manipulation compared to Python lists.
22. What are some common use cases for Seaborn?
  - Seaborn is used for visualizing distributions, relationships, and trends.
It's especially helpful in EDA using plots like heatmaps, boxplots, and pairplots.


```python
#1.How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(array_2d, axis=1)
print("1. Row-wise sum:\n", row_sums)


#2.Write a Pandas script to find the mean of a specific column in a DataFrame.
df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [85, 90, 95]})
mean_marks = df['Marks'].mean()
print("\n2. Mean of 'Marks' column:", mean_marks)

#3.Create a scatter plot using Matplotlib.
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
plt.scatter(x, y)
plt

#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
data = pd.DataFrame(np.random.rand(5, 3), columns=['A', 'B', 'C'])
corr = data.corr()
plt.figure()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("4. Correlation Matrix Heatmap")
plt.show()

#5.Generate a bar plot using Plotly.
bar_data = pd.DataFrame({'Fruit': ['Apple', 'Banana', 'Cherry'], 'Quantity': [10, 15, 7]})
fig = px.bar(bar_data, x='Fruit', y='Quantity', title="5. Fruit Quantity Bar Plot")
fig.show()


#6.Create a DataFrame and add a new column based on an existing column.
df2 = pd.DataFrame({'Salary': [30000, 45000, 50000]})
df2['Bonus'] = df2['Salary'] * 0.10
print("\n6. DataFrame with new 'Bonus' column:\n", df2)


#7.Write a program to perform element-wise multiplication of two NumPy arrays.
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
elementwise_mul = arr1 * arr2
print("\n7. Element-wise multiplication:", elementwise_mul)


#8.Create a line plot with multiple lines using Matplotlib.
x_vals = [0, 1, 2, 3]
y1 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]
plt.figure()
plt.plot(x_vals, y1, label='Line 1', marker='o')
plt.plot(x_vals, y2, label='Line 2', marker='x')
plt.title("8. Multiple Line Plot")
plt.legend()
plt.grid(True)
plt.show()


#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
df3 = pd.DataFrame({'ID': [1, 2, 3, 4], 'Score': [55, 70, 45, 90]})
filtered_df = df3[df3['Score'] > 60]
print("\n9. Filtered DataFrame (Score > 60):\n", filtered_df)


#10.Create a histogram using Seaborn to visualize a distribution.
data = np.random.normal(0, 1, 1000)
plt.figure()
sns.histplot(data, kde=True, bins=30)
plt.title("10. Histogram of Random Data")
plt


#11. Perform matrix multiplication using NumPy.
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
result = np.matmul(A, B)
print("\n11. Matrix Multiplication:\n", result)


#12.Use Pandas to load a CSV file and display its first 5 rows.
sample_csv = pd.DataFrame({'Name': ['X', 'Y', 'Z'], 'Age': [21, 22, 23]})
sample_csv.to_csv("sample.csv", index=False)
loaded_csv = pd.read_csv("sample.csv")
print("\n12. First 5 rows of loaded CSV:\n", loaded_csv.head())


#13.Create a 3D scatter plot using Plotly.
df4 = pd.DataFrame({
    'x': np.random.rand(10),
    'y': np.random.rand(10),
    'z': np.random.rand(10),
    'color': np.random.rand(10)
})
fig = px.scatter_3d(df4, x='x', y='y', z='z', color='color', title="13. 3D Scatter Plot")
fig.show()
```
